```python
# scrape amazon site pg 1,2,3 and 4 & print product name, price, ratings and no. of user reviews
import requests
from bs4 import BeautifulSoup

urls = [
    'https://www.amazon.in/s?i=electronics&rh=n%3A4363159031&fs=true&qid=1718168856&ref=sr_pg_1',
    'https://www.amazon.in/s?i=electronics&rh=n%3A4363159031&fs=true&page=2&qid=1718168805&ref=sr_pg_2',
    'https://www.amazon.in/s?i=electronics&rh=n%3A4363159031&fs=true&page=3&qid=1718168812&ref=sr_pg_3',
    'https://www.amazon.in/s?i=electronics&rh=n%3A4363159031&fs=true&page=4&qid=1718168834&ref=sr_pg_4'
]

headers = {
    'User-Agent': 'Mozilla/5.0 (X11; CrOS x86_64 14541.0.0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/125.0.0.0 Safari/537.36'
}

for url in urls:
    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')

        products = soup.find_all('div', {'data-component-type': 's-search-result'})

        for product in products:
            name_tag = product.find('h2', class_='a-size-mini a-spacing-none a-color-base s-line-clamp-2')
            if not name_tag:
                name_tag = product.find('span', class_='a-size-base-plus a-color-base a-text-normal')
            if name_tag:
                name = name_tag.text.strip()
            else:
                name = 'Name not found'

            price_whole = product.find('span', class_='a-price-whole')
            price_fraction = product.find('span', class_='a-price-fraction')
            if price_whole and price_fraction:
                price = f"{price_whole.text.strip()}{price_fraction.text.strip()}"
            elif price_whole:
                price = price_whole.text.strip()
            else:
                price = 'Price not found'

            reviews = product.find('span', class_='a-size-base')
            if reviews:
                reviews = reviews.text.strip()
            else:
                reviews = 'Reviews not found'

            rating = product.find('span', class_='a-icon-alt')
            if rating:
                rating = rating.text.strip()
            else:
                rating = 'Rating not found'

            print(f'Product Name: {name}')
            print(f'Price: {price}')
            print(f'Number of user reviews: {reviews}')
            print(f'Rating: {rating}')
            print('---' * 10)
    else:
        print(f'Failed to retrieve the page. Status code: {response.status_code}')
        print(f'URL: {url}')
        print('---' * 10)
```

```python
# convert the output to csv file
import csv
import requests
from bs4 import BeautifulSoup

urls = [
    'https://www.amazon.in/s?i=electronics&rh=n%3A4363159031&fs=true&qid=1718168856&ref=sr_pg_1',
    'https://www.amazon.in/s?i=electronics&rh=n%3A4363159031&fs=true&page=2&qid=1718168805&ref=sr_pg_2',
    'https://www.amazon.in/s?i=electronics&rh=n%3A4363159031&fs=true&page=3&qid=1718168812&ref=sr_pg_3',
    'https://www.amazon.in/s?i=electronics&rh=n%3A4363159031&fs=true&page=4&qid=1718168834&ref=sr_pg_4'
]

headers = {
    'User-Agent': 'Mozilla/5.0 (X11; CrOS x86_64 14541.0.0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/125.0.0.0 Safari/537.36'
}

with open('amazon_products.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Product Name', 'Price', 'Number of user reviews', 'Rating'])

    for url in urls:
        response = requests.get(url, headers=headers)

        if response.status_code == 200:
            soup = BeautifulSoup(response.content, 'html.parser')

            products = soup.find_all('div', {'data-component-type': 's-search-result'})

            for product in products:
                name_tag = product.find('h2', class_='a-size-mini a-spacing-none a-color-base s-line-clamp-2')
                if not name_tag:
                    name_tag = product.find('span', class_='a-size-base-plus a-color-base a-text-normal')
                if name_tag:
                    name = name_tag.text.strip()
                else:
                    name = 'Name not found'

                price_whole = product.find('span', class_='a-price-whole')
                price_fraction = product.find('span', class_='a-price-fraction')
                if price_whole and price_fraction:
                    price = f"{price_whole.text.strip()}{price_fraction.text.strip()}"
                elif price_whole:
                    price = price_whole.text.strip()
                else:
                    price = 'Price not found'

                reviews = product.find('span', class_='a-size-base')
                if reviews:
                    reviews = reviews.text.strip()
                else:
                    reviews = 'Reviews not found'

                rating = product.find('span', class_='a-icon-alt')
                if rating:
                    rating = rating.text.strip()
                else:
                    rating = 'Rating not found'

                writer.writerow([name, price, reviews, rating])

        else:
            print(f'Failed to retrieve the page. Status code: {response.status_code}')
            print(f'URL: {url}')
            print('---' * 10)
```

```python
# recommendation system based on the csv file
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv('/content/amazon_products.csv')

df['Rating'] = df['Rating'].str.extract('(\d+\.\d+|\d+)').astype(float)
df['Number of user reviews'] = df['Number of user reviews'].str.extract('(\d+)').astype(float)
df.dropna(inplace=True)

scaler = MinMaxScaler()
df[['Rating', 'Number of user reviews']] = scaler.fit_transform(df[['Rating', 'Number of user reviews']])

features = ['Rating', 'Number of user reviews']
similarity_matrix = cosine_similarity(df[features])

def recommend_smartphones(phone_title, num_recommendations=10):
    try:
        phone_index = df[df['Product Name'] == phone_title].index[0]
    except IndexError:
        print(f"Error: '{phone_title}' not found in the dataset.")
        return []

    similar_phones = sorted(list(enumerate(similarity_matrix[phone_index])), key=lambda x: x[1], reverse=True)
    similar_phones = similar_phones[1:num_recommendations+1]
    recommendations = [df.iloc[i[0]]['Product Name'] for i in similar_phones]
    return recommendations

recommendations = recommend_smartphones('Samsung Galaxy A15 5G (Light Blue, 6GB, 128GB Storage) | 50 MP Main Camera | Android 14 with One UI 6.0 | 12GB Expandable RAM | MediaTek Dimensity 6100+ | 5000 mAh Battery')
print("Recommended smartphones:")
for phone in recommendations:
    print(phone)

```
